```python
import os,sys
import pandas as pd
sys.path.append("../")
from utils.basic import get_files
```

```python
path = "../data/stocks/nyse"
files = get_files(path)
tickers = [f.split(".")[0] for f in files]
```

```python
first = True
for file in files:
    t = pd.read_csv(os.path.join(path,file),index_col=0)
    if first:
        df = pd.DataFrame(t["close"])
        df = df.rename(columns={"close":file.split(".")[0]})
        first=False
    else:
        df = df.join(t["close"],on="date") 
        df = df.rename(columns={"close":file.split(".")[0]})
df.head()
```
