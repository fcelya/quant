```python
import yfinance as yf
import pandas as pd
import requests
import os
```

```python
def get_stock_data(tickers, path, date_start=None, date_end=None, period=None):

    if not os.path.exists(path):
            os.makedirs(path)

    if date_start is not None and date_end is not None:
        for ticker in tickers:
            data = yf.download(tickers=ticker, start=date_start, end=date_end)
            data_path = os.path.join(path, ticker.upper()+".csv")
            data.columns = [c.lower() for c in data.columns]
            data.to_csv(data_path,index_label="date")
    elif period is not None:
        for ticker in tickers:
            data = yf.download(tickers=ticker, period=period)
            data_path = os.path.join(path, ticker.upper()+".csv")
            data.columns = [c.lower() for c in data.columns]
            data.to_csv(data_path,index_label="date")
    else:
        print("[WARNING] - Either start and end date or period must be specified")
```

```python
def save_download_file(path,url,name):
    if not os.path.exists(path):
            os.makedirs(path)
    req = requests.get(url)
    url_content = req.content
    file_path = os.path.join(path,name)
    with open(file_path,"wb") as f:
        f.write(url_content)
        f.close()
```

```python
sp = pd.read_csv("../data/tickers/sp500.csv")
splist = sp["Symbol"].to_list()
nyse = pd.read_csv("../data/tickers/nyse.csv")
nyselist = nyse["ACT Symbol"]

nyselist = [c.lower() for c in nyselist]
splist = [c.lower() for c in splist]

combinedlist = [c for c in splist if c in nyselist]
print(len(combinedlist))
```
