```python
def get_coin_historic(id,vs_currency,days,interval=None):
    """
    Get price, market cap and 24h volume (granularity auto)
    Granularity is automatic: 5m if <1d, 1h if 1-90d, 1d if >90d
    vs_currency: usd, eur, jpy, etc.
    days: 1,2,3,.....,max
    interval (optional): "daily" to force granularity to daily
    Coingecko is limited to 50 calls per minute
    """
    url = f"https://api.coingecko.com/api/v3/coins/{id}/market_chart"
    if interval:
        params = {
            "vs_currency":vs_currency,
            "days":days,
            "interval":interval
        }
    else:
        params = {
            "vs_currency":vs_currency,
            "days":days
        }


    resp = requests.get(url=url,params=params)
    data = resp.json() # Check the JSON Response Content documentation below
    d2df = {}
    d2df["date"] = [duple[0] for duple in data["prices"]]
    for key in data.keys():
        d2df[key] = [duple[1] for duple in data[key]]
    df = pd.DataFrame(d2df)
    df["date"] = pd.to_datetime(df["date"],unit="ms")
    df.index = df["date"]
    df = df.drop("date",axis=1)
    df.index = pd.to_datetime(df.index)
    return df
```
